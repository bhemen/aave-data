```python
from IPython.display import display, HTML
```

```python
pd.set_option('display.max_rows', None)  # Display all rows
pd.set_option('display.max_columns', None)  # Display all columns
pd.set_option('display.width', None)  # Use maximum width for display
pd.set_option('display.max_colwidth', None)  # Show the full content of each column
```

### Rows represent the chosen asset and columns represent the not chosen asset (i.e. it was present as collateral along with this but not chosen)

So an i,j th entry in the matrix would mean that --> (No. of times asset i was chosen and asset j not chosen) / (Total no. of times assets i and j were present together as collaterals)

```python
display(HTML(df.to_html()))
```

#### It also includes info on how often the chosen asset had a liquidation bonus greater than or equal to that of the not chosen asset. For instance 80% means that out of all instances when asset i was picked over asset j, 80% of times i had a liquidation bonus greater or equal to that of j.
